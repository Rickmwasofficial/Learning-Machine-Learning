## OO Plots for Pandas

```python
over_50 = heart_dis[heart_dis["age"] > 50]
over_50
```

```python
over_50.plot(kind="scatter", x="age", y="chol", c="target");
```

```python
# OO Method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter",
            y="chol",
            x="age",
            c="target",
            sharex=True,
            ax=ax);
#set limits
ax.set_xlim([45, 100])
```

## Subplot of chol, age, thalach

```python
plt.style.available
```

```python
plt.style.use("seaborn-v0_8")
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                               sharex=True)
scatter = ax1.scatter(x=over_50["age"],
            y=over_50["chol"],
            c=over_50["target"],
            cmap="winter");
scatter = ax2.scatter(x=over_50["age"],
            y=over_50["thalach"],
            c=over_50["target"],
            cmap="winter");

# Add Legends
ax1.legend(*scatter.legend_elements(), title="Target");
ax2.legend(*scatter.legend_elements(), title="Target");

#add mean lines
ax1.axhline(y=over_50["chol"].mean(),
            linestyle="--");
ax2.axhline(y=over_50["thalach"].mean(),
            linestyle="--");
# Add titles to the subplots
ax1.set(title="Relationship between cholestrol and heart disease",
        ylabel="Cholestrol"
        );

ax2.set(title="Relationship between Blood Pressure and heart disease",
        xlabel="Age",
        ylabel="Blood Pressure"
        );

ax1.set_xlim([45, 100]);
ax2.set_ylim([60, 200]);
# Add title to the figure
fig.suptitle("Relationship between age, cholestrol and blood pressure with Heart Disease", fontsize=14, fontweight="bold");
```
